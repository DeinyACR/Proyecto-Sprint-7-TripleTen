## Limpieza de datos 

Para que la app que se desarrollara traaje de una manera mas optima y sea mas entendible la informacion que se proyectara en los graficos se haran las siguientes modificaciones a los datos:

- En la columna 'price' y 'odometer':
    - Para mejor legibilidad las cantidades de separaran por comas.
- En la columna 'paint_color':
    - Se remplazaran los datos NaN por 'unknown'
- En la columna 'is_4wd':
    - Se remplazaran los 1 por 'SI' y los NaN por 'NO'
- En la columna 'date_posted':
    - Se convertiran los datos tipo object a datetime.

    


 ### Columna price

```python
car_data['price'] = car_data['price'].apply(lambda x: f'{x:,}')
car_data['odometer'] = car_data['odometer'].apply(lambda x: f'{x:,}')
car_data
```

```python
fig = px.histogram(car_data, x="odometer") # crear un histograma
fig.show() # crear gráfico de dispersión
```
