# Analisis exploratorio de los datos

En este documento se encuentra la exploracion de los datos y diseño de algunos graficos con el fin de comprender mas a fondo nuestos datos y trabajar de la mejor mandera posible. 

```python
import pandas as pd
import plotly.express as px

car_data = pd.read_csv('../vehicles_us.csv') # leer los datos
car_data

```

```python
car_data['date_posted'] = pd.to_datetime(car_data['date_posted'])
year = car_data['date_posted'].value_counts().sort_index().reset_index()
fig = px.line(year, x='date_posted', y='count', title = 'Número de anuncios al día')
fig.show()
```

```python
print(car_data['condition'].describe())
```

```python
condition = car_data['condition'].value_counts().reset_index()
fig = px.pie(condition, values = 'count', names = 'condition', title = 'Condición de los carros')
fig.show()
```

```python
table = car_data.pivot_table(index = 'condition',
                             columns = 'fuel', 
                             values = 'model',
                             aggfunc = 'count')
fig = px.bar(table, x= table.index, y = table.columns, color="fuel", title="Clasificacion de carros por condicion y tipo de combustible")
fig.show()
```
